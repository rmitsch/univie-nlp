#### Exercise 7

Find all the four-letter words in text6. With the help of a frequency distribution
(FreqDist), show these words in decreasing order of frequency.

````python
# Get frequency distribution.
fdist = FreqDist(text6)

# Filter FreqDist by length of words.
words = [(w, fdist[w]) for w in set(text6) if len(w) == 4]

# Sort words by occurence.
words.sort(key=lambda tup: tup[1], reverse=True)

words
````
