### Accessing Text Corpora and Lexical Resources

```python
import matplotlib.pyplot as plt
import nltk
from nltk.book import *
import nltk.corpus
from nltk.corpus import wordnet as wn
```

### Exercise 1

Plot the conditional frequency distribution of how the usage of the words "men", "women",
and "people" has changed over time in the Inaugural Address Corpus.

```python
plt.figure(figsize=(20, 5))
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['men', 'women', 'people']
    if w.lower().startswith(target)
)
cfd.plot()
```

### Exercise 2

Plot a conditional frequency distribution over the Names Corpus that allows you to see
which initial letters are more frequent for males vs. females.

```python
plt.figure(figsize=(20, 5))

cfd = nltk.ConditionalFreqDist(
    (target, w[0])
    for target in nltk.corpus.names.fileids()
    for w in nltk.corpus.names.words(target)
)
cfd.plot()
```

### Exercise 3

What percentage of noun synsets have no hyponyms? You can get all noun synsets using
wn.all synsets('n').

```python
synsets_wo_hyponyms = [synset for synset in wn.all_synsets('n') if len(synset.hyponyms()) == 0]
len(synsets_wo_hyponyms) / len(list(wn.all_synsets('n')))
```
